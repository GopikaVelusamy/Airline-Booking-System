```python
import pandas as pd
import random
import string
```

```python
# Global variables
df1 = pd.DataFrame()
df2 = pd.DataFrame()
df3 = pd.DataFrame()
customer_detail = []
```

```python
def airline_details():
    global df1
    n1 = int(input("Enter the number of airlines available: "))
    airlines = [input(f"Enter the airline name {i+1}: ") for i in range(n1)]
    print("Airlines:", airlines)

    n2 = int(input("Enter the number of classes: "))
    classes = [input(f"Enter the class name {j+1}: ") for j in range(n2)]
    print("Classes:", classes)

    tickets = []
    for airline in airlines:
        for cls in classes:
            total_tickets = int(input(f"Enter total number of tickets for {airline} in {cls}: "))
            tickets.append({'Airline': airline, 'Class': cls, 'Tickets': total_tickets})

    df1 = pd.DataFrame(tickets)
    print(df1)
```

```python
def book_ticket():
    global df1, df2, df3, customer_detail

    # Collect customer details
    name = input("Enter the customer name: ")
    check_name=input("check if the entered name is correct? (Yes/No):")
    if check_name=='No':
      name=input("Enter Your name properly:")
    age = input("Enter the customer's age: ")
    check_age=input("check if the entered age is correct? (Yes/No):")
    if check_name=='No':
      age=input("Enter Your age properly:")
    gender = input("Enter the customer's gender: ")
    check_gender=input("check if the entered gender is correct? (Yes/No):")
    if check_gender=='No':
      gender=input("Enter Your gender properly:")
    airline = input("Enter the airline the customer wants to book: ")
    check_airline=input("check if the entered airline is correct? (Yes/No):")
    if check_airline=='No':
      airline=input("Enter Your airline properly:")
    clas = input("Enter the class the customer wants to book: ")
    check_clas=input("check if the entered class is correct? (Yes/No):")
    if check_clas=='No':
      clas=input("Enter Your class properly:")

    # Confirm booking
    confirmation = input("Do you want to confirm your booking? (Yes/No): ")
    if confirmation.lower() != 'yes':
        print("Your booking has been cancelled.")
        return

    # Generate ticket number
    valid_chars = string.digits
    ticket_number = ''.join(random.choices(valid_chars, k=7))

    # Update airline details
    ticket_index = df1[(df1['Airline'] == airline) & (df1['Class'] == clas)].index
    if not ticket_index.empty and df1.loc[ticket_index, 'Tickets'].values[0] > 0:
        df1.loc[ticket_index, 'Tickets'] -= 1
        print("Updated airline details:")
        print(df1)

        # Add customer to the booking list
        customer = {'Name': name, 'Age': age, 'Gender': gender, 'Airline': airline, 'Class': clas, 'Ticket Number': ticket_number}
        customer_detail.append(customer)
        df2 = pd.DataFrame(customer_detail)

        # Update customer details after booking
        global df3
        df3 = df2.copy()

        print("Congratulations, your tickets are confirmed.")
        print("Customer details with ticket number:")
        print(df3)
    else:
        print(f"No available tickets for {airline} {clas}. Booking failed for {name}.")
```

```python
def print_all_customers_after_booking():
    global df3
    if not df3.empty:
        print("All customer details after bookings:")
        print(df3)
        df3.to_csv('/content/Passenger.csv')
    else:
        print("No customer details available.")
```

```python
def find_customer_by_ticket(ticket_number):
    global df3
    customer = df3[df3['Ticket Number'] == ticket_number]
    if not customer.empty:
        print(customer)
    else:
        print("No customer found with the given ticket number.")
```

```python
def cancellation_ticket():
    global df1, df2, df3, customer_detail
    cancellation = input("Do you want to cancel your ticket? (Yes/No): ")
    if cancellation.lower() == 'yes':
        ticket_no = input("Enter your ticket number: ")
        check_Ticket_no=input("check whether the entered ticket number is correct? (Yes/No):")
        if check_Ticket_no=='No':
          ticket_no = input("Enter your ticket number:")
        if ticket_no in df3['Ticket Number'].values:
            customer_info = df3[df3['Ticket Number'] == ticket_no].iloc[0]
            df3 = df3[df3['Ticket Number'] != ticket_no]
            customer_detail = [customer for customer in customer_detail if customer['Ticket Number'] != ticket_no]
            airline = customer_info['Airline']
            clas = customer_info['Class']
            ticket_index = df1[(df1['Airline'] == airline) & (df1['Class'] == clas)].index
            df1.loc[ticket_index, 'Tickets'] += 1
            print("Updated airline details after cancellation:")
            print(df1)
            print("Updated customer details after cancellation:")
            print(df3)
        else:
            print("The given ticket number is not in the list.")
    else:
        print("Your ticket is still confirmed.")
```

```python
def Flight_details():
    global df1, df3
    flights = df1['Airline'].unique()  # Get unique airlines from df1

    for flight in flights:
        passengers = df3[df3['Airline'] == flight]  # Get passengers for the flight
        if passengers.empty:
            print(f"There are no passengers in the flight {flight}")
        else:
            print(f"Passengers in the flight {flight}:")
            print(passengers)
            print("-" * 40)
            a=df3[df3['Airline']==flight]
            a.to_csv(f'{flight}_passengers.csv')
```

```python
def menu():
    print('Menu:')
    print('1. Airline details')
    print('2. Ticket Booking')
    print('3. All customers details after booking')
    print('4. Extract customer details')
    print('5. Cancellation of tickets')
    print('6. Flight details')
```

```python

while True:
    menu()
    choice = int(input("Enter the choice: "))
    if choice == 1:
        airline_details()
    elif choice == 2:
        book_ticket()
    elif choice == 3:
        print_all_customers_after_booking()
    elif choice == 4:
        ticket_number = input("Enter the ticket number: ")
        find_customer_by_ticket(ticket_number)
    elif choice == 5:
        cancellation_ticket()
    elif choice == 6:
        Flight_details()
    else:
        print("Invalid choice. Please try again.")
        break
```
